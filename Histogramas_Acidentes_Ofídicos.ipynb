```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados da planilha "Cobras 2020"
file_path = '/content/Casos e óbitos.xlsx'
df_cobras = pd.read_excel(file_path, sheet_name='Cobras 2022')

# Definir a largura das barras e o número de barras
bar_width = 0.6
index = df_cobras.index

# Criar o gráfico de barras para os casos
fig, ax1 = plt.subplots(figsize=(12, 8))
bars = ax1.bar(index, df_cobras['Casos'], bar_width, color='blue', edgecolor='black', label='Casos')

# Adicionar os valores no topo das barras
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=13)

# Criar os pontos para os óbitos, sem as linhas
ax2 = ax1.twinx()
ax2.scatter(index, df_cobras['Óbitos'], color='red', label='Óbitos', zorder=5)

# Adicionar os valores ao lado dos pontos de óbitos
for i, txt in enumerate(df_cobras['Óbitos']):
    ax2.text(index[i] + 0.25, df_cobras['Óbitos'][i], int(txt), ha='left', va='center', fontsize=13, color='red')

# Configurar os eixos
ax1.set_xlabel('Espécies')
ax1.set_ylabel('Número de Casos', color='blue', fontsize=13)
ax2.set_ylabel('Número de Óbitos', color='red', fontsize=13)

# Definir os rótulos das barras
ax1.set_xticks(index)
ax1.set_xticklabels(df_cobras['Espécie'], rotation=45, ha='right')

# Adicionar as legendas
fig.legend(loc='lower right', bbox_to_anchor=(1, 0.05), frameon=False)

# Ajustar o título e mostrar o gráfico
plt.title('Número de Casos e Óbitos por Espécie (2022)')
plt.show()
```
